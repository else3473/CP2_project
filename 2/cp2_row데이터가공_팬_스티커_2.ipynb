```python
import pandas as pd
```

```python
df_fan_sticker=pd.read_csv('cp2_팬채팅_스티커_데이터.csv')
```

```python
df_fan_sticker
```

```python
df_fan_sticker.dropna(subset='user_nickname', inplace=True)
```

```python
#닉네임 기준 중복값 확인
df_fan_sticker[df_fan_sticker.user_nickname.duplicated(keep=False)]
```

```python
#가장 최근까지 유지된 채팅방을 남기고 삭제 (이탈고객을 찾는 프로젝트이므로 최근데이터가 중요)
#37 89 92 110 122 139 152 인덱스는 먼저 지우고, 나머지는 keep first
df_fan_sticker = df_fan_sticker.drop([37, 89, 92, 110, 122, 139, 152])
df_fan_sticker = df_fan_sticker.drop_duplicates(subset='user_nickname')
df_fan_sticker
```
