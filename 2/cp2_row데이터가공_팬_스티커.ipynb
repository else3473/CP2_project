### df_fan에서 채팅최초일자, 채팅최근일자, 채팅기간(day/week), 누적채팅량, 최근 일자기준 1주일간 채팅량 구하기

```python
df_fan
```

```python
#df_tmp1 채팅방 별 가장 최근 채팅일자
df_tmp1 = df_fan.groupby('chat_room_id',as_index=False).max()[['chat_room_id','time']]
#날짜만 가져오기
df_tmp1['time'] = df_tmp1['time'].apply(lambda x :x.date)
#컬럼이름을 last_chat_date로 변경
df_tmp1.rename(columns = {'time':'last_fan_chat_date'}, inplace=True)
df_tmp1
```

```python
#df_tmp2 채팅방 별 가장 최초 채팅일자(기간중)
df_tmp2 = df_fan.groupby('chat_room_id',as_index=False).min()[['chat_room_id','time']]
#날짜만 가져오기
df_tmp2['time'] = df_tmp2['time'].apply(lambda x :x.date)
#컬럼이름을 last_chat_date로 변경
df_tmp2.rename(columns = {'time':'first_fan_chat_date'}, inplace=True)
df_tmp2
```

```python
#df_tmp3 채팅방 팬 채팅 누적 횟수
df_tmp3 = df_fan.groupby('chat_room_id',as_index=False).count()[['chat_room_id','time']]
#컬럼이름을 chat_count로 변경
df_tmp3.rename(columns = {'time':'fan_chat_count'}, inplace=True)
df_tmp3
```

```python
df_tmp1 = df_tmp1.merge(df_tmp2,on='chat_room_id')
df_tmp1['fan_chat_period']=df_tmp1['last_fan_chat_date']-df_tmp1['first_fan_chat_date']
df_tmp1['fan_chat_period'] = df_tmp1['fan_chat_period'].dt.days+1
df_tmp1
```

```python
#날짜로 몇주인지 구하는 함수
def div_week(num):
    if num <6:
        return 1
    elif num % 7 == 0:
        return int(num / 7)
    else:
        return int(num // 7+1)
```

```python
df_tmp1['fan_chat_week'] = df_tmp1['fan_chat_period'].apply(div_week)
df_tmp1
```

```python
#df_tmp3도 합침
df_tmp1 = df_tmp1.merge(df_tmp3,on='chat_room_id')
df_tmp1
```

```python
df_tmp1['avg_fan_chat'] = round ( df_tmp1['fan_chat_count'] /  df_tmp1['fan_chat_week'] , 2)
df_tmp1
```

```python
from datetime import datetime, timedelta
```

```python
df_fan_date = df_fan.copy()
df_fan_date['time'] = df_fan['time'].apply(lambda x :x.date)
```

```python
tmp_list = []
for idx, row in df_tmp1.iterrows():
    day = row["last_fan_chat_date"]+timedelta(days=-7)
    room = row["chat_room_id"]
    tmp_list.append( df_fan_date[ (df_fan_date['chat_room_id'] ==room) & (df_fan_date['time'] > day)].shape[0] )
    
tmp_list
```

```python
df_tmp1['last_week_fan_chat_count'] = tmp_list
df_tmp1
```

### df_sticker에서  땡큐스티커 횟수, 땡큐스티커 평균금액 , 땡큐스티커 누적금액 구하기

```python
#문자열을 숫자로 바꿈
df_sticker['price'] = df_sticker['product_price_string'].astype(int)
df_sticker
```

```python
#df_tmp4 스티커 누적 횟수
df_tmp4 = df_sticker.groupby('chat_room_id',as_index=False).count()[['chat_room_id','time']]
#컬럼이름을 sticker_count로 변경
df_tmp4.rename(columns = {'time':'sticker_count'}, inplace=True)
df_tmp4
```

```python
#df_tmp4 스티커 누적 금액
df_tmp5 = df_sticker.groupby('chat_room_id',as_index=False).sum()[['chat_room_id','price']]
#컬럼이름을 sticker_count로 변경
df_tmp5.rename(columns = {'price':'Total_sticker_sales'}, inplace=True)
df_tmp5
```

```python
df_tmp4 = df_tmp4.merge(df_tmp5, on='chat_room_id')
df_tmp4
```

```python
df_tmp4['avg_sticker_price'] = round ( df_tmp4['Total_sticker_sales'] / df_tmp4['sticker_count'] ,1)
df_tmp4
```
